```python
name_data_file = 'hipermercado.db'
complete_path = path + '/02_Data/01_Originals/' + name_data_file
```

```python
# !pip install --upgrade sqlalchemy pandas
import sqlalchemy as sa
```

```python
con = sa.create_engine('sqlite:///' + complete_path)
```

```python
from sqlalchemy import inspect
insp = inspect(con)
tables = insp.get_table_names()
tables
```

```python
cal = pd.read_sql('calendar', con)
cal.drop(columns='index',inplace=True)
cal.head()
```

```python
sales = pd.read_sql('sales', con)
sales.drop(columns='index',inplace=True)
sales.head()
```

```python
prices = pd.read_sql('sell_prices', con)
prices.drop(columns='index',inplace=True)
prices.head()
```

### Union of calendar with sales

```python
sales.head()
```

```python
sales = sales.melt(id_vars = sales.columns[0:6],
                   var_name = 'd',
                   value_name = 'sales')
sales
```

```python
sales.drop(columns='id',inplace=True)
```

#### Cross with calendar to enter your information

```python
df = sales.merge(right = cal, how = 'left', on = 'd')
df.head()
```

### Incorporation of prices

```python
prices.head()
```

```python
df = df.merge(right = prices, how = 'left', on = ['store_id','item_id','wm_yr_wk'])
df
```

```python
df[['store_id','item_id','wm_yr_wk','d','sell_price']]\
    .sort_values(by = ['store_id','item_id','wm_yr_wk','d'])\
    .head(10)
```

### Reorder and put date as index

```python
df.head(1)
```

```python
order = ['date',
         'state_id',
         'store_id',
         'dept_id',
         'cat_id',
         'item_id',
         'wm_yr_wk',
         'd',
         'sales',
         'sell_price',
         'year',
         'month',
         'wday',
         'weekday',
         'event_name_1',
         'event_type_1',
         'event_name_2',
         'event_type_2']
```

```python
df = df[order].set_index('date')
df
```
